```python
import matplotlib.pyplot as plt
import numpy as np


# Data
models = ['DT', 'DT + Smote', 'DT + Smote + RFE', 'ANN Smote + RFE']
accuracy = [0.7462, 0.5847, 0.7125, 0.4907]
auc = [0.6886, 0.5847, 0.65, 0.5938]

# labels
def add_labels(axes, values):
    for i, value in enumerate(values):
        axes.text(i, value + 0.01, f"{value:.2%}", ha='center', va='bottom')

fig, axes = plt.subplots(1, 2, figsize=(13, 6))

axes[0].bar(models, accuracy, color=['blue', 'green', 'red'])
axes[0].set_title('Model Accuracy Comparison')
axes[0].set_ylabel('Accuracy')
axes[0].set_ylim(0, max(accuracy) + 0.1) 
add_labels(axes[0], accuracy)


axes[1].bar(models, auc, color=['blue', 'green', 'red'])
axes[1].set_title('Model AUC Comparison')
axes[1].set_ylabel('AUC')
axes[1].set_ylim(0, max(auc) + 0.1) 
add_labels(axes[1], auc)

plt.tight_layout()
plt.show()

```
